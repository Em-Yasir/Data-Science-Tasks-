#### Task 1: Merging 12 months of sales data into a single file csv file.

```python
directory_path = r'E:\Data Analysis\Python Practice\Data Science Tasks in Python\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data'

# Get a list of all files in the directory
file_list = os.listdir(directory_path)

all_months_data = pd.DataFrame()

for file in file_list:
    file_path = os.path.join(directory_path, file)
    df = pd.read_csv(file_path)
    all_months_data = pd.concat([all_months_data, df])

# Reset the index of the concatenated DataFrame
all_months_data.reset_index(drop=True, inplace=True)

# Downloading a new file
all_months_data.to_csv(r'E:\Data Analysis\Python Practice\Data Science Tasks in Python\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data\all_data.csv',index = False)
```

```python
#another method of creating a new column from an existing column. is by using .apply funtion with lambda

#.apply method is not very efficient when you are dealing with a very big data
def get_city (address):
    return address.split(',')[1]
def get_state (address):
    return address.split(',')[2].split(' ')[1]

all_data.loc[:,'CITY'] = all_data['Purchase Address'].apply(lambda x: get_city(x)+ ' '+ get_state(x))

#all_data.drop(columns='City', inplace = True)
#all_data.drop(columns='Street_Address', inplace = True)
#all_data.drop(columns='Zip_Code', inplace = True)
all_data
```

### Question no 4: What products are most often sold together?


```python
duplicated_orderid = all_data[all_data['Order ID'].duplicated(keep = False)]

# use transform method to join values from two rows into a single row 

duplicated_orderid["Grouped"] = duplicated_orderid.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

duplicated_orderid = duplicated_orderid[['Order ID', 'Grouped']].drop_duplicates()

duplicated_orderid
```

```python
#The combinations function from itertools is used to generate all possible combinations of a given iterable (like a list or string) of a specified length.

from itertools import combinations 

#The Counter class from collections is used to count the occurrences of elements in a collection (like a list or string). It returns a dictionary where the keys are the elements and the values are their respective counts. 

from collections import Counter
```

```python
count = Counter()

for row in duplicated_orderid["Grouped"]:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

count.most_common(10)
                
```

### Question no 5: What product sold the most and why do you think it sold the most?

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum(numeric_only = True)['Quantity Ordered']

products = [product for product , df in product_group]

plt.bar(products, quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Products')
plt.xticks(products, rotation='vertical', size = 8)
plt.show()
```
